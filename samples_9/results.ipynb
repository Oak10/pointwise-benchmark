```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick  # for percentage matplotlib tick format

# Intervals for the error rate
interval_labels = [
    '[0.00, 0.10]', ']0.10, 0.20]', ']0.20, 0.30]', ']0.30, 0.40]', ']0.40, 0.50]',
    ']0.50, 0.60]', ']0.60, 0.70]', ']0.70, 0.80]', ']0.80, 0.90]', ']0.90, 1.00]'
]

def plot_boxplot_intervals_error_rate(all_intervals_error_0, all_intervals_error_1, plot_label):
    positions_0 = np.arange(len(interval_labels)) - 0.2
    positions_1 = np.arange(len(interval_labels)) + 0.2

    plt.figure(figsize=(13, 5))

    # Boxplot for Label 0
    plt.boxplot(all_intervals_error_0, positions=positions_0, widths=0.35,
                patch_artist=True,
                boxprops=dict(facecolor='lightblue'),
                medianprops=dict(color='red'))

    # Boxplot for Label 1
    plt.boxplot(all_intervals_error_1, positions=positions_1, widths=0.35,
                patch_artist=True,
                boxprops=dict(facecolor='lightgreen'),
                medianprops=dict(color='red'))

    plt.xticks(np.arange(len(interval_labels)), interval_labels)
    plt.xlabel('Reliability Intervals')
    plt.ylabel('Error Rate (%)')
    plt.title(f'Boxplot of Error Rates Across Intervals ({plot_label} - Label 0 vs Label 1)')
    

    #  percentage
    plt.yticks(np.arange(0, 1.05, .05))
    plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0))

    plt.grid(True)

    # Legend 
    blue_patch = plt.Line2D([0], [0], color='lightblue', lw=4, label='Label 0')
    green_patch = plt.Line2D([0], [0], color='lightgreen', lw=4, label='Label 1')
    plt.legend(handles=[blue_patch, green_patch], loc='upper left')
    plt.show()


def calculate_error_rate(correct, incorrect):
    error_rates = []
    for c, i in zip(correct, incorrect):
        total = c + i
        error_rate = i / total if total != 0 else 0
        error_rates.append(error_rate)
    return error_rates
```
